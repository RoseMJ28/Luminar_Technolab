# **Functions**

```python
#arange()

a=np.arange(10)
print(a)
b=np.arange(1,10,2)
b
```

```python
type(a)
```

```python
#reshape()

c=a.reshape(2,5)
print(c)
d=a.reshape(5,2,1)
print(d)
```

```python
#create a 1D array contains all the multiples of 3 upto 36 using arange()
#create all possible 2D and 3D arrays using reshape()


f=np.arange(3,37,3)
f
s=f.reshape(4,3)
print(s)
r=f.reshape(3,4)
print(r)
t=f.reshape(2,3,2)
print(t)
u=f.reshape(2,2,3)
print(u)
v=f.reshape(3,2,2)
print(v)
```

```python
#flatten()--> converting to 1 dimension
f.flatten()
```

```python
s.max()
print(s.max(axis=1))  #--> max in each row
print(s.max(axis=0))  #--> max in each col
print(s.min())
print(s.min(axis=1))
print(s.min(axis=0))
```

```python
# argmax()  --> return the index count of max value element
print(s.argmax())
print(s.argmax(axis=1)) # returns index of max value in each row
print(s.argmax(axis=0))
```

```python
print(s.argmin())
print(s.argmin(axis=1))
print(s.argmin(axis=0))
```
