# **Combining Datasets**

---



```python
import pandas as pd
df1=pd.DataFrame({'RollNo':[23,34,45,56],
                  'Name':['Arun','Deepak','Vinu','Jasmine'],
                  'class':[12,12,11,12]})
df1
```

```python
df2=pd.DataFrame({'RollNo':[12,37],
                  'Name':['Aiswarya','Athira'],
                  'class':[11,11]})
df2
```

```python
df3=pd.DataFrame({'Mark':[34,35,45,55],
                  'Gender':['M','M','M','F']})
df3
```

```python
df4=pd.DataFrame({'RollNo':[23,35,45,55],
                  'Mark':[34,35,45,55],
                  'Gender':['M','M','M','F']})
df4
```

```python
pd.concat([df1,df2])   # combines row-wise,without mentioning axis
```

```python
pd.concat([df1,df2],ignore_index=True)   # combines row-wise,with mentioning ignore_index
```

```python
pd.concat([df1,df3],axis=1)   # combines column-wise
```

```python
df1.merge(df4,how='inner',on='RollNo')  # common observation only
```

```python
df1.merge(df4,how='left',on='RollNo')  # left observation
```

```python
df1.merge(df4,how='right',on='RollNo')  # right observation
```

```python
df1.merge(df4,how='outer',on='RollNo')  # all observation
```
