```python
import pandas as pd
d=pd.read_csv('/content/drive/MyDrive/Datasets/train-data.csv')
d
```

```python
# drop newprice column
d.drop(['New_Price','Unnamed: 0'],axis=1,inplace=True)
d
```

```python
d.isna().sum()
```

```python
d.dtypes
```

ml algos can't be applied on datas with dtype 'object'

```python
d['Mileage']=d['Mileage'].str.replace('km/kg','')
d['Mileage']=d['Mileage'].str.replace('kmpl','')
d['Engine']=d['Engine'].str.replace('CC','')
d['Power']=d['Power'].str.replace('bhp','')
d
```

```python
d.dtypes
```

```python
d['Engine']=d['Engine'].astype(float)
d['Mileage']=d['Mileage'].astype(float)
d['Power']=d['Power'].str.replace('null','0.0')
d['Power']=d['Power'].astype(float)
d.dtypes
```

```python
d['Mileage'].fillna(d['Mileage'].mean(),inplace=True)
d['Engine'].fillna(d['Engine'].mode()[0],inplace=True)
d['Power'].fillna(d['Power'].mean(),inplace=True)
d['Seats'].fillna(d['Seats'].mode()[0],inplace=True)
```
