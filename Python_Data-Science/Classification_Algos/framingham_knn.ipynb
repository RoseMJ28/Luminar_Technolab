```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
d=pd.read_csv('/content/drive/MyDrive/Datasets/framingham.csv')
d
```

```python
d.isna().sum()
```

```python
d.dtypes
```

```python
d.drop(['education','glucose'],axis=1,inplace=True)
```

```python
d['cigsPerDay'].fillna(d['cigsPerDay'].mode()[0],inplace=True)
d['BPMeds'].fillna(d['BPMeds'].mode()[0],inplace=True)
d['totChol'].fillna(d['totChol'].mean(),inplace=True)
d['BMI'].fillna(d['BMI'].mean(),inplace=True)
d['heartRate'].fillna(d['heartRate'].mean(),inplace=True)
d
```

```python
X=d.iloc[:,:-1]
y=d.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaler=scaler.fit_transform(X)
X_scaler
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaler,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```
