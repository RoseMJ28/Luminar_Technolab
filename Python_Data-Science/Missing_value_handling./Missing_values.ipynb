# **Missing value Handling**


1.   Drop rows having missing values
2.   Drop the column if it contains so many missing values
3.   Fill with a random value
4.   Filling with mean values
5.   Filling with median
6.   Filling with mode

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/missing_data.csv')
df
```

Drop rows having missing values

```python
df.isna().sum()  # to find the missing values
```

```python
df1=df.copy()
df1.dropna(inplace=True)  # drop rows with missing values
df1.reset_index(drop=True,inplace=True)
df1
```

fillna()   --> fills the missing value with random value

```python
df2=df.copy()
# df2.fillna(100,inplace=True)
df2['Calories'].fillna(100,inplace=True)
df2['Date'].fillna('2022-01-01',inplace=True)

# OR

df2['Calories']=df2['Calories'].fillna(100)
df2['Date']=df2['Date'].fillna('2022-01-01')

df2
```

filling with mean

```python
df3=df.copy()
mn=df3['Calories'].mean()
df3['Calories'].fillna(mn,inplace=True)
df3
```

fill with median

```python
df4=df.copy()
md=df4['Calories'].median()
df4['Calories'].fillna(md,inplace=True)
df4
```

filling with  mode

```python
df5=df.copy()
mo=df5['Calories'].mode()[0]  # for getting single value
df5['Calories'].fillna(mo,inplace=True)
df5
```
