```python
X=d.iloc[:,:-1]
y=d.iloc[:,-1]
```

there is onlu one column so scaling is not need if done no problem

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
y_test.values
```

# to get actual and predicted values

```python
result=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
result
```

```python
lr.intercept_   # to get y intercept i.e 'c' in equation y=mx+c
```

```python
lr.coef_    # to get slope in equation y=mx+c
```

for visualisation

```python
import matplotlib.pyplot as plt
plt.xlabel('TV marketing')
plt.ylabel('Sales')
plt.scatter(X_train,y_train)
```

```python
plt.xlabel('TV marketing')
plt.ylabel('Sales')
plt.scatter(X_test,y_test)   #actual value
plt.scatter(X_test,y_pred,c='red')  # predicted value
```

# performance measures of regression



1.   r2 score  =1-(rss/tss)

2.   mean absolute error




```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
print('R2_score:',r2_score(y_test,y_pred))
print('mean absolute error:',mean_absolute_error(y_test,y_pred))
print('mean squared error:',mean_squared_error(y_test,y_pred))
print('root mean squared error:',root_mean_squared_error(y_test,y_pred))
```
