```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
df
```

```python
df1=df[['total_bill','tip','size']]
df1
```

```python
df1.corr()  # to show how these columns are related to each other
```

```python
sns.heatmap(df1.corr(),annot=True) # we can plot the co relation, annot=True shows the values also, highest value is a feature that depend on target
```

# Jointplot

```python
sns.jointplot(data=df1,x='total_bill',y='tip')
```

```python
sns.jointplot(data=df,x='size',y='tip',hue='day') # hue --use categorical values
```

# Pairplot

```python
sns.pairplot(df)
```

# Countplot

```python
sns.countplot(data=df,x='sex',hue='smoker')
```

# Barplot

```python
sns.barplot(data=df,x='size',y='total_bill')  # line indicates dispersion
```

# Distribution plot

```python
sns.displot(df['tip'])
```

# Boxplot

```python
sns.boxplot(data=df,x='day',y='total_bill')  #line  inside box indicates median(also called 50 percentile,Q2),min value(25 percentile,Q1) thats down and max (75 percentile,Q3) is upper line,all these lines are in box
# Outliers/Anomaly are athe other lines and bubbles outside the box ,which are found in respect to IQR
# Q3 + 1.5 * IQR , Q1 - 1.5 * IQR
#  IQR--> inter quartile range (width of box)
```

# Violin plot

```python
sns.violinplot(data=df,x='day',y='total_bill')
```
