```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/movies_cleaned (1).csv',names=['ID','Movies','Year','Rating','Duration'])
df
```

```python
# 1.Give column name as Id, Movie, Year, Rating, Duration
```

```python
# 2.Display the number of rows and number of columns
df.shape
```

```python
# 3.Display the five observations that have highest duration
df.loc[df['Duration']].sort_values(by='Duration',ascending=False).head()
```

```python
# 4.Remove if any duplicate exists and change the index accordingly
df.drop(df.columns[[0]], axis=1,inplace=True)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
# 5.Rename the column name ‘Movie’ as Film
df.rename(columns={'Movies':'Film'},inplace=True)
```

```python
# 6.Display all the observation starting from index 100 upto 200
df.iloc[100:201]
```

```python
# 7.Display the name and releasing year of the film with maximum duration
df.loc[df['Duration']].sort_values(by='Duration',ascending=True).head(1)[['Film','Year']]
```

```python
# 8.Display the count of films released in each year.

df['Year'].value_counts()

# OR

df.groupby('Year')['Film'].count()
```

```python
# 9.Find the film released between 1986-1996 having highest duration
df.loc[(df['Year']>=1986) & (df['Year']<=1996)].sort_values(by='Duration',ascending=False).head(1)[['Film','Duration']]

```

```python
# 10.Find the films having duration greater than 7000 and rating above 3.5
df.loc[(df['Duration']>7000) & (df['Rating']>3.5)]
```

```python
# 11.Print the details about 5 films having less rating , display in the descending order
df.sort_values(by='Rating',ascending=False).head()
```

```python
# 12.Print the name of the film released in 1996 have highest rating
df.loc[(df['Year']==1996)].sort_values(by='Rating',ascending=False).head(1)[['Film']]
```

```python
# Find highest duration of films released in each year
df.groupby('Year')['Duration'].max()
```

```python
df.groupby('Year')['Duration'].min()
```

```python
# find the count of film with same releasing year and same rating
df.groupby(['Year','Rating'])['Film'].count()
```
