```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/drive/MyDrive/Datasets/train-data.csv')
df
```

```python
df.isna().sum()
```

```python
df['Brand']=df['Name'].apply(lambda x:x.split()[0])
```

```python
df.drop('New_Price',inplace=True,axis=1)
df.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
df.drop('Name',inplace=True,axis=1)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df
```

```python
df['Mileage']=df['Mileage'].str.replace('km/kg','')
df['Mileage']=df['Mileage'].str.replace('kmpl','')
df['Engine']=df['Engine'].str.replace('CC','')
df['Power']=df['Power'].str.replace('bhp','')
```

```python
df['Engine']=df['Engine'].astype(float)
df['Mileage']=df['Mileage'].astype(float)
df['Power']=df['Power'].str.replace('null','0.0')
df['Power']=df['Power'].astype(float)
```

```python
df['Mileage'].fillna(df['Mileage'].mean(),inplace=True)
df['Engine'].fillna(df['Engine'].mode()[0],inplace=True)
df.loc[df.Power==0.0,'Power']=np.NaN
df['Power'].fillna(df['Power'].mean(),inplace=True)
df['Seats'].fillna(df['Seats'].mode()[0],inplace=True)
```

labelencoder used to transform categorical data to numerical data used instead of map()

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Location']=encoder.fit_transform(df['Location'])
df['Fuel_Type']=encoder.fit_transform(df['Fuel_Type'])
df['Transmission']=encoder.fit_transform(df['Transmission'])
df['Owner_Type']=encoder.fit_transform(df['Owner_Type'])
df['Brand']=encoder.fit_transform(df['Brand'])
df

```

```python
X=df.drop(['Price'],axis=1)
y=df.iloc[:,-2]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
```

```python
y_pred=mlr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
print('R2 score:',r2_score(y_test,y_pred)*100)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
knr.fit(X_train,y_train)
```

```python
y_pred1=knr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print('R2 score:',r2_score(y_test,y_pred1)*100)
```
