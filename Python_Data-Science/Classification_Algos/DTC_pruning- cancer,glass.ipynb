```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/cancer_data.csv')
df
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
#
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',random_state=1,max_depth=6)
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
dt.score(X_test,y_test)  #can be used before y_pred prediction score() directly predict and compare with y_test
```

```python
dt.score(X_train,y_train)
```
