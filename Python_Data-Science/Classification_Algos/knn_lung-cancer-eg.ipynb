```python
import pandas as pd
d=pd.read_csv('/content/drive/MyDrive/Datasets/lung_cancer_examples.csv')
d
```

```python
# drop the columns
d1=d.copy()
d1.drop(['Name','Surname'],axis=1,inplace=True)
d1
```

```python
d1.isna().sum()
```

```python
d1.dtypes
```

```python
X=d1.iloc[:,:-1] # X beacuse multiple cols are taken
X  # features
```

```python
y=d1.iloc[:,-1]
y  # target
```

# Preprocessing the file

 Here  each column is haviing diff ranges so we use scaling,we are using MinMax Scalar[0-1]

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)  # transform() used to scale the X
X_scaled
```

splitting the dataset into 70% for training and 30% for testing... so we split dataset accordingly

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

X_train and y_train needs to have same shape,X_test and y_test needs to have same shape........ after splitting

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
# model created after training
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

testing X_test and stored in y_pred, y_test is used for testing the predicted values in y_pred

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
# predicting the accuracy of y_pred with y_test
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
y_new=knn.predict(scaler.transform([[34,5,3,4]]))
if y_new==0:
  print('Not suffering Lung cancer')
else:
  print('Suffering lung cancer')
```
