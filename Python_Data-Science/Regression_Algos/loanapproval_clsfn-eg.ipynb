```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/LoanApprovalPrediction.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

string values can be handled by mode()

```python
df['Gender'].unique()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

converting object datatype to numeric using map()

```python
df['Gender']=df['Gender'].map({'Male':1,'Female':0})
df['Married']=df['Married'].map({'Yes':1,'No':0})
df['Dependents']=df['Dependents'].map({'0':0,'1':1,'2':2,'3+':3})
df['Education']=df['Education'].map({'Graduate':1,'Not Graduate':0})
df['Self_Employed']=df['Self_Employed'].map({'Yes':1,'No':0})
df['Property_Area']=df['Property_Area'].map({'Urban':2,'Rural':0,'Semiurban':1})
df['Loan_Status']=df['Loan_Status'].map({'Y':1,'N':0})
```

```python
df
```

```python
X=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
y_new=knn.predict([[1,0,0,1,0,5849,0.0,146.412162,360.0,1,0]])
y_new
```

```python
if y_new==1:
  print('Loan approved')
else:
  print('Loan not approved')
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf')
svc.fit(X_train,y_train)
```

```python
y_pred1=svc.predict(X_test)
y_pred1
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred1)*100)
```
